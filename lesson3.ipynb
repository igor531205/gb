*Импортируем библиотеки pandas и numpy*

```python
import pandas as pd
import numpy as np
```

*Загрузим "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn*

```python
from sklearn.datasets import load_boston

# Загрузка данных
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

*Загрузим "California housing dataset" из встроенных наборов данных библиотеки sklearn*

```python
from sklearn.datasets import fetch_california_housing

# Загрузка данных
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
df.head()
```

*Разделим данные на тренировочные и тестовые выборки*

```python
from sklearn.model_selection import train_test_split

# Разделение данных на признаки и целевую переменную
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

*Создадим модель линейной регрессии и обучим еe*

```python
from sklearn.linear_model import LinearRegression

# Создание модели линейной регрессии
lr = LinearRegression()

# Обучение модели на тренировочных данных
lr.fit(X_train, y_train)

# Предсказания на тестовой выборке
y_pred = lr.predict(X_test)
```

*Оценка модели*

```python
from sklearn.metrics import mean_squared_error, r2_score

# Вычисление метрик
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

*Загрузим датасет и создадим датафрейм*

```python
# Загрузка датасета
df = pd.read_csv('creditcard.csv')

# Настройка отображения всех столбцов
pd.options.display.max_columns = 100;

# Удаление строк с пропущенными значениями
df = df.dropna()

# Просмотр первых 10 строк датафрейма
df.head(10)
```

*Подготовка данных*

```python
# Разделение данных на признаки и целевую переменную
X = df.drop('Class', axis=1)
y = df['Class']

# Разделение данных на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

*Настройка и обучение модели с помощью GridSearchCV*

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Параметры для поиска по сетке
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]

# Создание GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                           param_grid=parameters,
                           scoring='roc_auc',
                           cv=3)

# Обучение на тренировочных данных
grid_search.fit(X_train, y_train)

# Лучшие параметры
best_params = grid_search.best_params_
print(f'Best parameters: {best_params}')
```

*Предсказание и вычисление AUC*

```python
from sklearn.metrics import roc_auc_score

# Предсказание вероятностей классов
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]

# Вычисление AUC на тестовых данных
auc_test = roc_auc_score(y_test, y_pred_proba)
print(f'Test AUC: {auc_test}')
```
